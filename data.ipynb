```python
import pandas as pd
import numpy as np
from scipy.io import savemat
```

```python
metabolomics = pd.read_csv("IMOVNN_data/metabolomics.csv", index_col=0)
```

```python
metagenomics = pd.read_csv("IMOVNN_data/metagenomics.csv", index_col=0)
```

```python
metagenomics
```

```python
metatranscriptomics = pd.read_csv("IMOVNN_data/metatranscriptomics.csv", index_col=0)
```

```python
metadata = pd.read_csv('IMOVNN_data/metadata.csv', index_col=0)
```

```python
metatranscriptomics
```

```python
def log(base,x):
    return np.log(x)/np.log(base)
metagenomics=log(2,2*metagenomics+0.00001)
```

```python
metagenomics
```

```python
metagenomics_external_id = metagenomics.index.tolist()
metabolomics_external_id = metabolomics.index.tolist()
metatranscriptomics_external_id = metatranscriptomics.index.tolist()
metadata_external_id = metadata.index.tolist()
```

```python
available_omics = np.zeros((2891, 3))
for i in range(len(metadata_external_id)):
    if metadata_external_id[i] in metagenomics_external_id:
        available_omics[i, 0] = 1
    if metadata_external_id[i] in metabolomics_external_id:
        available_omics[i, 1] = 1
    if metadata_external_id[i] in metatranscriptomics_external_id:
        available_omics[i, 2] = 1
print(available_omics)
```

```python
savemat("available_omics.mat", {"available_omics": available_omics})
```

```python
num_features = 578

# Identify missing values
metagenomics_missing_values = [id_ for id_ in metadata_external_id if id_ not in metagenomics.index]

# Create a DataFrame for the missing values with all features set to zero
metagenomics_missing_df = pd.DataFrame(
    np.zeros((len(metagenomics_missing_values), num_features)), 
    index=metagenomics_missing_values
)

# Ensure the columns of the missing_df match the original DataFrame
metagenomics_missing_df.columns = metagenomics.columns
metagenomics = pd.concat([metagenomics, metagenomics_missing_df])


print(metagenomics)
```

```python
metagenomics = metagenomics.reindex(metadata_external_id)
print(metagenomics)
```

```python
num_features = 596

# Identify missing values
metabolomics_missing_values = [id_ for id_ in metadata_external_id if id_ not in metabolomics.index]

# Create a DataFrame for the missing values with all features set to zero
metabolomics_missing_df = pd.DataFrame(
    np.zeros((len(metabolomics_missing_values), num_features)), 
    index=metabolomics_missing_values
)

# Ensure the columns of the missing_df match the original DataFrame
metabolomics_missing_df.columns = metabolomics.columns
metabolomics = pd.concat([metabolomics, metabolomics_missing_df])
metabolomics = metabolomics.reindex(metadata_external_id)


print(metabolomics)
```

```python
num_features = 421

# Identify missing values
metatranscriptomics_missing_values = [id_ for id_ in metadata_external_id if id_ not in metatranscriptomics.index]

# Create a DataFrame for the missing values with all features set to zero
metatranscriptomics_missing_df = pd.DataFrame(
    np.zeros((len(metatranscriptomics_missing_values), num_features)), 
    index=metatranscriptomics_missing_values
)

# Ensure the columns of the missing_df match the original DataFrame
metatranscriptomics_missing_df.columns = metatranscriptomics.columns
metatranscriptomics = pd.concat([metatranscriptomics, metatranscriptomics_missing_df])
metatranscriptomics = metatranscriptomics.reindex(metadata_external_id)


print(metatranscriptomics)
```

```python
metatranscriptomics
```

```python
def normalize_column(column):
    return (column - column.min()) / (column.max() - column.min())

# Apply normalization to each column
metagenomics = metagenomics.apply(normalize_column)
metabolomics = metabolomics.apply(normalize_column)
metatranscriptomics = metatranscriptomics.apply(normalize_column)
```

```python
metabolomics.applymap(lambda x: x != 0 and not pd.isna(x)).sum()
```

```python
metagenomics.fillna(0, inplace=True)
metabolomics.fillna(0, inplace=True)
metatranscriptomics.fillna(0, inplace=True)
```

```python
arr1 = metagenomics.to_numpy()
arr2 = metabolomics.to_numpy()
arr3 = metatranscriptomics.to_numpy()
```

```python
savemat('metagenomics_data.mat', {'metagenomics_X': arr1})
savemat('metabolomics_data.mat', {'metabolomics_X': arr2})
savemat('metatranscriptomics_data.mat', {'metatranscriptomics_X': arr2})
```
